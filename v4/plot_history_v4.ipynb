# v4: Train prior on all data, train posterior separately on each domain

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_history(history):
    epochs = np.arange(len(history))+1
    plt.plot(epochs, history['val_loss'], 'b', epochs, history['loss'], 'b--', 
             epochs, history['val_acc'], 'r', epochs, history['acc'], 'r--')
    plt.ylim([0,1])
    plt.legend(['loss (testing)', 'loss (training)', 'accuracy (testing)', 'accuracy (training)'], bbox_to_anchor=(1.55, 1))
    plt.show()
    print('Maximum test accuracy is %0.4f at epoch %i' %(max(history.val_acc), np.argmax(history.val_acc)+1))
```

### Setup:
prior: 
  * optimizer=RMSProp(lr=1e-3, epsilon=1e-6)
  
posterior:
  * optimizer=RMSProp(lr=1e-3, epsilon=1e-6)

```python
datapath = '/home/lorenzo/PycharmProjects/domainadaption/sorted_data_acl/'
categories = ['books', 'dvd', 'electronics', 'kitchen']
setting = 'globalEmbeddings_traintestsplit_shuffled'

print 'prior: all'
prior_history = pd.read_csv(datapath + 'all/v4/keras_' + setting + '/keras_history.csv')
plot_history(prior_history)

print '\n\posterior'
for category in categories:
    print '\ncategory: ' + category
    posterior_history = pd.read_csv(datapath + category +'/v4/keras_' + setting + '/keras_history.csv')
    plot_history(posterior_history)
```
