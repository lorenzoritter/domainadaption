# v5: Use Embedding layer

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_history(history):
    epochs = np.arange(len(history))+1
    plt.plot(epochs, history['val_loss'], 'b', epochs, history['loss'], 'b--', 
             epochs, history['val_acc'], 'r', epochs, history['acc'], 'r--')
    plt.ylim([0,1])
    plt.legend(['loss (testing)', 'loss (training)', 'accuracy (testing)', 'accuracy (training)'], bbox_to_anchor=(1.55, 1))
    plt.show()
    print('Maximum test accuracy is %0.4f at epoch %i' %(max(history.val_acc), np.argmax(history.val_acc)+1))
```

### Setup:
keras 1.0.3

prior: 
  * optimizer=RMSProp(lr=1e-3, epsilon=1e-8) # default
  * Embedding layer static
  
posterior:
  * optimizer=RMSProp(lr=1e-3, epsilon=1e-8) # default
  * Embedding layer trainable, no Dropout

```python
datapath = '/home/lorenzo/PycharmProjects/domainadaption/sorted_data_acl/'
categories = ['books', 'dvd', 'electronics', 'kitchen']
setting_prior = ''
setting_posterior = 'noEmbeddingDropout'

print 'prior: all'
prior_history = pd.read_csv(datapath + 'all/v5/keras_' + setting_prior + '/keras_history.csv')
plot_history(prior_history)

print '\nposterior'
for category in categories:
    print '\ncategory: ' + category
    posterior_history = pd.read_csv(datapath + category +'/v5/keras_' + setting_posterior + '/keras_history.csv')
    plot_history(posterior_history)
    
import evaluate_prior_posterior_v5
evaluate_prior_posterior_v5.evaluate_prior(setting_prior)
evaluate_prior_posterior_v5.evaluate_posterior(setting_posterior)
```
