```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
datapath = '/home/lorenzo/PycharmProjects/domainadaption/sorted_data_acl/'
categories = ['books', 'dvd', 'electronics', 'kitchen_&_housewares', 'all']
```

### books
1600 train, 400 test <br>
100 epochs <br>
max_length = 200, embedding_dim = 50

```python
history = pd.read_csv('/home/lorenzo/PycharmProjects/domainadaption/sorted_data_acl/books/keras_history.csv')
epochs = np.arange(len(history))+1
```

```python
plt.plot(epochs, history['val_loss'], 'b', epochs, history['loss'], 'b--', 
         epochs, history['val_acc'], 'r', epochs, history['acc'], 'r--')
plt.ylim([0,1])
plt.legend(['loss (testing)', 'loss (training)', 'accuracy (testing)', 'accuracy (training)'], bbox_to_anchor=(1.55, 1))
plt.show()
print('Maximum test accuracy is %0.4f at epoch %i' %(max(history.val_acc), np.argmax(history.val_acc)+1))
```

### all categories
1000 epochs <br>
max_length = 200, embedding_dim = 250<br>

LSTM Regularization: input=0.5, recurrent=0.3

```python
settings = 'embedDim250_FullRegularization_1000eps'

for category in categories:
    print '\ncategory: ' + category
    history = pd.read_csv(datapath + category +'/keras_' + settings + '/keras_history.csv')
    epochs = np.arange(len(history))+1
    
    plt.plot(epochs, history['val_loss'], 'b', epochs, history['loss'], 'b--', 
         epochs, history['val_acc'], 'r', epochs, history['acc'], 'r--')
    plt.ylim([0,1])
    plt.legend(['loss (testing)', 'loss (training)', 'accuracy (testing)', 'accuracy (training)'], bbox_to_anchor=(1.55, 1))
    plt.show()
    print('Maximum test accuracy is %0.4f at epoch %i' %(max(history.val_acc), np.argmax(history.val_acc)+1))
```
